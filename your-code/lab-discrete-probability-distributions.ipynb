```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#number of apples and oranges
num_apples = 60
num_oranges = 40
total_fruits = num_apples + num_oranges
# probabilities calculation 
prob_apple = num_apples/total_fruits
prob_orange = num_oranges/total_fruits

print(f"probability of picking an apple: {prob_apple}")
print(f"probability of picking an orange: {prob_orange}")
```

```python
#probabilities
prob_apple = 0.6
prob_orange = 0.4

#probabilities calculation
prob_first_5_apples = prob_apple ** 5
prob_first_5_apples_next_15_oranges = (prob_apple ** 5) * (prob_orange ** 15)

print (f"probability that the first 5 friuits are all apples: {prob_first_5_apples}")
print (f"probability that the first 5 fruits are all apples and the next 15 fruits are all oranges oranges: {prob_first_5_apples_next_15_oranges}")
```

```python
## Binomial Distribution
import math
n = 20
k = 5
p = 0.6

#binomial coefficient
binom_coeff = math.comb (n, k)

#probability calculations
probability = binom_coeff * (p ** k) * ((1 - p) ** (n - k))

print (f"probability of getting exactly 5 apples and 15 oranges in a sample of 20 fruits: {probability:.10f}")
```

```python
#cumulative distribution function
from scipy.stats import binom
n = 20
p = 0.6

# calculating the cumulative probability of getting less than 5 apples
probabilty = binom.cdf(4, n, p)
print (f"probability of getting less than 5 apples in a sample of 20 fruits: {probability:.10f}")
```

```python
#PDF
import numpy as np
x = np.arange(0, n + 1)
pmf = binom.pmf (x, n, p)
plt.figure(figsize=(10, 6))
plt.bar(x, pmf, color = 'blue', alpha =0.7)
plt.xlabel('number of apples picked')
plt.ylabel('probability')
plt.title('binomial distribution pmf(n=20, p=0.6)')
plt.xticks(x)
plt.grid(True)
plt.show()
```

```python
#Poisson Distributio
lambda_ = 2.3
k = 5
probability = (lambda_ ** k * math.exp(lambda_))/ math.factorial(k)
print (f"probability of scoring exactly {k} goals in a match: {probability:.10f}")
```

```python
lambda_ = 5.3
k_values = range (11)
probabilities = [(lambda_ ** k * math.exp(-lambda_))/ math.factorial(k) for k in k_values]

plt.figure(figsize = (10, 6))
plt.bar(k_values, probabilities, color ='blue', alpha=0.7)
plt.xlabel('Number of goals')
plt.ylabel(probabilities)
plt.title('poisson distribution PMF for Number of Goals')
plt.xticks (k_values)
plt.grid(True)
plt.show()
```
